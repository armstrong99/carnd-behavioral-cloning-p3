```python
import numpy as np
def _get_translated_image(img, y, trans_range):
    rows, cols, _ = img.shape
    tr_x = trans_range * np.random.uniform() - trans_range/2
    y = y + tr_x/trans_range * 2 *.4
    tr_y = 10 * np.random.uniform() - 10/2
    Trans_M = np.float32([[1,0, tr_x], [0,1, tr_y]])
    img = cv2.warpAffine(img, Trans_M, (cols, rows))
    return img, y
```

```python
image = cv2.imread("drive_data/IMG/center_2017_10_30_20_32_08_352.jpg")
angle = -0.3012811
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
image, y = _get_translated_image(image, angle, 100)
```

```python
image = cv2.imread("drive_data/IMG/center_2017_10_30_20_32_08_352.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
angle = -0.3012811
image, y = _get_translated_image(image, angle, 100)
def _get_brightnessed_image(img):
    img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    random_bright = .25 + np.random.uniform()
    img[:,:,2] = img[:,:,2] * random_bright
    img = cv2.cvtColor(img, cv2.COLOR_HSV2RGB)
    return img

```

```python
img = _get_brightnessed_image(image)
plt.imshow(img)
plt.show()
```
